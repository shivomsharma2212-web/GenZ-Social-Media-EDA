```python
import pandas as pd
import matplotlib.pyplot as plt

#  Gen Z Social Media Usage Analysis

file_path = "data/genz_social_media_usage_1M.csv"

df = pd.read_csv(file_path)

# STEP 2 : Read Data

print("\n========== FIRST 5 ROWS ==========")
print(df.head())

print("\n========== DATASET SHAPE ==========")
print(df.shape)

print("\n========== COLUMN NAMES ==========")
print(df.columns)

print("\n========== DATASET INFO ==========")
print(df.info())

print("\n========== MISSING VALUES ==========")
print(df.isnull().sum())

print("\n========== DESCRIPTIVE STATISTICS ==========")
print(df.describe())


# STEP 3: Gender Analysis

print("\n========== GENDER ANALYSIS ==========")
gender_counts = df['gender'].value_counts()
print(gender_counts)

plt.figure(figsize=(6,5))
gender_counts.plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,6))
gender_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Percentage Distribution")
plt.ylabel("")
plt.tight_layout()
plt.show()

# STEP 4: Age Analysis

print("\n========== AGE ANALYSIS ==========")
print(df['age'].describe())

plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=15)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Users")
plt.tight_layout()
plt.show()


# STEP 5: Platform Analysis

print("\n========== PRIMARY PLATFORM ANALYSIS ==========")
platform_counts = df['primary_platform'].value_counts()
print(platform_counts)

plt.figure(figsize=(8,5))
platform_counts.plot(kind='bar')
plt.title("Most Used Social Media Platform")
plt.xlabel("Platform")
plt.ylabel("Users")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# STEP 6: Daily Usage Analysis

print("\n========== DAILY USAGE HOURS ==========")
print(df['daily_usage_hours'].describe())

plt.figure(figsize=(8,5))
plt.hist(df['daily_usage_hours'], bins=20)
plt.title("Daily Usage Hours Distribution")
plt.xlabel("Hours")
plt.ylabel("Users")
plt.tight_layout()
plt.show()

# STEP 7: Addiction Level Analysis

print("\n========== ADDICTION LEVEL ANALYSIS ==========")
addiction_counts = df['addiction_level'].value_counts()
print(addiction_counts)

plt.figure(figsize=(8,5))
addiction_counts.plot(kind='bar')
plt.title("Addiction Level Distribution")
plt.xlabel("Addiction Level")
plt.ylabel("Users")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


# STEP 8: Mental Health Analysis

print("\n========== MENTAL HEALTH SCORE ==========")
print(df['mental_health_score'].describe())

plt.figure(figsize=(8,5))
plt.hist(df['mental_health_score'], bins=20)
plt.title("Mental Health Score Distribution")
plt.xlabel("Mental Health Score")
plt.ylabel("Users")
plt.tight_layout()
plt.show()


# STEP 9: Age vs Daily Usage

plt.figure(figsize=(8,5))
plt.scatter(df['age'], df['daily_usage_hours'])
plt.title("Age vs Daily Usage Hours")
plt.xlabel("Age")
plt.ylabel("Daily Usage Hours")
plt.tight_layout()
plt.show()


# STEP 10: Daily Usage vs Mental Health

plt.figure(figsize=(8,5))
plt.scatter(df['daily_usage_hours'], df['mental_health_score'])
plt.title("Daily Usage Hours vs Mental Health Score")
plt.xlabel("Daily Usage Hours")
plt.ylabel("Mental Health Score")
plt.tight_layout()
plt.show()


# STEP 11: Correlation Analysis

print("\n========== CORRELATION MATRIX ==========")

numeric_df = df.select_dtypes(include=['int64', 'float64'])
correlation = numeric_df.corr()
print(correlation)

plt.figure(figsize=(10,8))
plt.imshow(correlation, aspect='auto')
plt.colorbar()
plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=90)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()


# FINAL NOTE

print("\nEDA Complete Successfully")
print("Now write observations + insights + conclusions for each graph.")
```
